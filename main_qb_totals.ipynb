2022 WR Projections

```python
X_train2022, X_test2022 = train_test('qb', num_years=5, year_for=2022, totals=True)
X_train2022.head(10)
```

```python
xgbPPR_results = revised_run(X_train2022, X_test2022, 'qb', year_for='2022', totals=True)
xgbPPR_results.head(50)
```

```python
enetPPR_results = revised_run(X_train2022, X_test2022, 'qb', model_ = 'enet', pca=False, totals=True)
enetPPR_results.head(50)
```

```python
rfPPR_results = revised_run(X_train2022, X_test2022, 'qb', model_ = 'rf', pca=False, totals=True)
rfPPR_results.head(50)
```

```python
svrPPR_results = revised_run(X_train2022, X_test2022, 'qb', model_ = 'svr', totals=True)
svrPPR_results.head(50)
```

```python
mean_results_PPR = xgbPPR_results.append([rfPPR_results, svrPPR_results, enetPPR_results], ignore_index = True)[['name', 'proj fpts']].groupby('name', as_index=False).mean().sort_values('proj fpts', ascending=False)
mean_results_PPR = mean_results_PPR.reset_index(drop=True)
mean_results_PPR['rank'] = mean_results_PPR.index + 1
mean_results_PPR.to_csv('projections/qb_PPR_2022_mean_totals.csv', index = False)
mean_results_PPR.head(50)
```
