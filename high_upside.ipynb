```python
from scrape import predict_elite
from collections import defaultdict
```

```python
times = defaultdict(int)
for mod in ['rf', 'kn', 'lr', 'svc', 'xgb']:
    if mod in ['kn', 'svc']:
        pca = True
    else:
        pca = False
    result = predict_elite('rb', mod, 0.87, pca=pca, year_for=2022)
    best = result[result['class'] == 0]['name']
    for name in best:
        times[name] += 1
print(times)
```
